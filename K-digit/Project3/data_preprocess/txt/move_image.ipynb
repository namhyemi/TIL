```python
import os
import shutil
import pandas as pd


## 폴더 생성
def createFolder(directory):
    try:
        if not os.path.exists(directory): ## 이미 존재하는지 체크
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

# filename = './check_image.csv'
filename = './check_image_training.csv' 
df = pd.read_csv(filename, index_col=0)

error_count = 0
success_count = 0

for i in range(len(df['file'])):
    
    check_jpg = '.jpg'
    check_jpeg = '.jpeg'
    check_JPG = '.JPG'
    check_JPEG = '.JPEG'

    file = df['file'][i] ## 파일 이름
    file = file.split('.')[0] ## 확장자 제거
    folder = file.split('_')[2] ## 파일 해당 폴더 

#     oPath = './../bpm/dataset/training/' + folder
#     nPath = './../bpm/dataset/training/temp/' + folder
    oPath = './training/' + folder
    nPath = './training/training_temp/' + folder

    createFolder(nPath) ## 폴더 생성

    for i in ['.jpg','.jpeg','.JPG','.JPEG']: ## 해당 txt 이름을 가진 4종류의 확장자 모두 찾아보기 
        try:
            shutil.move(oPath+'/'+file+ i, nPath+'/'+file+ i)
            success_count += 1
            print(success_count, file+'.jpg')
            break ## 확장자 하나라도 발견하면 for문 종료
        except FileNotFoundError:
            if i == '.jpg': check_jpg = file
            elif i == '.jpeg': check_jpeg = file
            elif i == '.JPG' : check_JPG = file
            elif i == '.JPGE' : check_JPEG = file
#             print('FileNotFoundError :', file)
 
    if(check_jpg == check_jpeg == check_JPG == check_JPEG):
        error_count += 1
        print('FileNotFoundError :', file) ## 4종류로 모두 못 찾은 파일명 출력
    
        
print('파일 이동 성공 : ', success_count)       
print('파일 이동 실패 : ', error_count)
```
