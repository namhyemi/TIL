```python
import os
import shutil
import pandas as pd


##  
def createFolder(directory):
    try:
        if not os.path.exists(directory): ##   
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

# filename = './check_image.csv'
filename = './check_image_training.csv' 
df = pd.read_csv(filename, index_col=0)

error_count = 0
success_count = 0

for i in range(len(df['file'])):
    
    check_jpg = '.jpg'
    check_jpeg = '.jpeg'
    check_JPG = '.JPG'
    check_JPEG = '.JPEG'

    file = df['file'][i] ##  
    file = file.split('.')[0] ##  
    folder = file.split('_')[2] ##    

#     oPath = './../bpm/dataset/training/' + folder
#     nPath = './../bpm/dataset/training/temp/' + folder
    oPath = './training/' + folder
    nPath = './training/training_temp/' + folder

    createFolder(nPath) ##  

    for i in ['.jpg','.jpeg','.JPG','.JPEG']: ##  txt   4    
        try:
            shutil.move(oPath+'/'+file+ i, nPath+'/'+file+ i)
            success_count += 1
            print(success_count, file+'.jpg')
            break ##    for 
        except FileNotFoundError:
            if i == '.jpg': check_jpg = file
            elif i == '.jpeg': check_jpeg = file
            elif i == '.JPG' : check_JPG = file
            elif i == '.JPGE' : check_JPEG = file
#             print('FileNotFoundError :', file)
 
    if(check_jpg == check_jpeg == check_JPG == check_JPEG):
        error_count += 1
        print('FileNotFoundError :', file) ## 4     
    
        
print('   : ', success_count)       
print('   : ', error_count)
```
