```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from matplotlib import font_manager, rc
font_path = './malgun.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
reviews1_df = pd.read_csv('List_star1_1.csv' , encoding='utf-8', names=['word', 'count'])
reviews1_df.head()
```

```python
plt.figure(figsize=(20,10))
plt.title('별점 1점 (상위 20단어)', pad=10)
plt.ylabel('사용 횟수')
plt.xlabel('단어')
plt.bar(reviews1_df['word'][:20], reviews1_df['count'][:20], width=0.5)

plt.show()
```
