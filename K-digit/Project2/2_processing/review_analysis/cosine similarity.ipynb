```python
restaurant_df['store_code'] = restaurant_df['place_code']
restaurant_df.drop(['place_code'], axis=1, inplace=True)
restaurant_df.head()
```

```python
reviews_df = pd.read_csv('restaurant_df.csv', encoding='utf-8', index_col=0)
reviews_df.head()
```

```python
join_df = pd.merge(left=restaurant_df, right=reviews_df, how='inner', on=['store_code', 'ItemID'])
join_df.head()
```

```python
join_df = join_df.dropna(axis=0)
```

```python
count_df = join_df.groupby('store_code').size().reset_index(name = 'review_count')
count_df
```

```python
join_df.groupby('store_code')['Rating'].sum()
```

```python
test_df = pd.merge(left=join_df, right=count_df, how='inner', on='store_code')
test_df.head()
```

```python
test_df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer  # 피체 벡터화
from sklearn.metrics.pairwise import cosine_similarity  # 코사인 유사도
```

```python
count_vect_category = CountVectorizer(min_df=0, ngram_range=(1,2))
place_category = count_vect_category.fit_transform(test_df['category']) 

place_simi_cate = cosine_similarity(place_category, place_category) 
place_simi_cate_sorted_ind = place_simi_cate.argsort()[:, ::-1]
```
